# Import Libraries

```python
import pickle
import pandas as pd
import numpy as np
import random
# import seaborn as sns
# from sklearn.decomposition import PCA
import joblib
```

# Load Model

```python
best_model = joblib.load('final_model.pkl')
```

# Inference

```python
# Menentukan nilai seed untuk memastikan hasil yang konsisten
seed = 123
random.seed(seed)
np.random.seed(seed)

# Jumlah data yang akan dibuat
jumlah_data = 500

# Membuat data acak untuk masing-masing fitur
tenure = np.random.uniform(low=0, high=50, size=jumlah_data)
jumlah_device = random.choices([1, 2, 3, 4, 5, 6], k=jumlah_data)
skor_kepuasan = random.choices([1, 2, 3, 4, 5], k=jumlah_data)
ada_keluhan = random.choices([0, 1], k=jumlah_data)
hari_sejak_order_terakhir = np.random.uniform(0, 30, jumlah_data)
jumlah_cashback = np.random.uniform(0, 300, jumlah_data)
kategori_order = random.choices(['Laptop & Accessory', 'Mobile Phone', 'Others', 'Fashion', 'Grocery'], k=jumlah_data)
status_pernikahan = random.choices(['Single', 'Divorced', 'Married'], k=jumlah_data)

# Menggabungkan semua fitur ke dalam satu DataFrame
data_simulasi = pd.DataFrame({
    'Tenure': tenure,
    'NumberOfDeviceRegistered': jumlah_device,
    'SatisfactionScore': skor_kepuasan,
    'Complain': ada_keluhan,
    'DaySinceLastOrder': hari_sejak_order_terakhir,
    'CashbackAmount': jumlah_cashback,
    'PreferedOrderCat': kategori_order,
    'MaritalStatus': status_pernikahan
})
```

```python
y_inf_pred = pd.DataFrame(best_model.predict(data_simulasi))
y_inf_pred.value_counts()
```

```python
# Menyisipkan hasil prediksi ke dalam DataFrame
data_simulasi['Churn_Predicted'] = best_model.predict(data_simulasi)
```
